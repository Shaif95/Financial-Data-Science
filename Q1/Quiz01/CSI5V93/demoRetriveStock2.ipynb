# We visualize data a little bit

```python
# we calculate log return for the stock and put it as a variable in DataFrame

goog['UP_DOWN']=goog["Close"]-goog["Open"]>0
goog['LogReturn'] = np.log(goog['Close']/goog['Close'].shift(1))
goog.head()
```

# Write a visualization module to do 2x2 subplots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')

def plot_data(data):

  fig=plt.figure(figsize=(12,8))

  plt.subplot(2,2,1)
  plt.plot(data['Close'], c='orange')
  plt.xlabel('year')
  plt.ylabel('price')
  plt.grid('on')
  
  plt.subplot(2,2,2)
  sns.kdeplot(data['Close'], color='b')
  plt.xlabel('price')
  plt.ylabel('density')
  
  plt.subplot(2,2,3)
  sns.boxplot(data['Close'], color='gold')


  plt.subplot(2,2,4)
  sns.boxplot(data['LogReturn'], color='c')


  plt.show()
```

```python
plot_data(goog)
```

```python
sns.pairplot(data=goog, hue='UP_DOWN', markers=['o', 's'])
plt.show()
```
